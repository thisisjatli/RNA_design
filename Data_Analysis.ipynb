```python
file_path = "taro_results.pt"
taro_data = torch.load(file_path)
```

```python
taro_df = pd.DataFrame(taro_data)
```

```python
taro_df = taro_df.transpose() 
taro_df
```

```python
taro_df["length"] = taro_df["seq"].apply(len)
taro_df
```

```python
taro_df = taro_df.astype({
    'recovery_avg': float,
    'runtime' : float
})

correlation_recovery_avg = taro_df["length"].corr(taro_df["recovery_avg"])

plt.figure(figsize=(8, 6))
sns.scatterplot(x="length", y="recovery_avg", data=taro_df, color="blue", s=50)
sns.regplot(x="length", y="recovery_avg", data=taro_df, scatter=False, color="red", label=f"Correlation = {correlation_recovery_avg:.2f}")
plt.title("Correlation between Recovery and Length", fontsize=16)
plt.xlabel("Length", fontsize=12)
plt.ylabel("Recovery", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()

```

```python
test_dic = {} # test data 的基本資料(based on taro_results)
test_seq = taro_df["seq"].unique().tolist()
for seq in test_seq:
    if seq in data:
        test_dic[seq] = data[seq]
```

```python
len(test_dic)
```

```python
for key, value in test_dic.items():
    print(value['type_list'])
```

```python
taro_df_new = taro_df[taro_df['seq'].isin(test_dic.keys())]
```

```python
taro_df_new.shape
test_dic[taro_df_new.iloc[0]['seq']]['type_list'][0]
```

```python
def fun1(x):
    # print(test_dic[x]['type_list'][0])
    return test_dic[x]['type_list'][0]
    

taro_df_new["structure_type"] = taro_df_new['seq'].apply(fun1)
taro_df_new
```

```python
# Group by 'structure_type' and calculate the mean of 'recovery rate'
grouped = taro_df_new.groupby('structure_type')['recovery_avg'].describe()

# Show the summary statistics
print(grouped)
```

```python
# Create a box plot to compare 'recovery_rate' across 'structure_type'
plt.figure(figsize=(10, 6))
sns.boxplot(x='structure_type', y='recovery_avg', data=taro_df_new)
plt.title("Recovery Rate by Structure Type")
plt.xlabel("Structure Type")
plt.ylabel("Recovery Rate")
plt.show()
```

```python
for key, value in test_dic.items():
    print(value['rfam_list'])
```

```python
def fun2(x):
    return test_dic[x]['rfam_list'][0]
    

taro_df_new["RFAM"] = taro_df_new['seq'].apply(fun2)


# Group by 'RFAM' and calculate the mean of 'recovery rate'
grouped = taro_df_new.groupby('RFAM')['recovery_avg'].describe()

# Show the summary statistics
print(grouped)

# Create a box plot to compare 'recovery_rate' across 'RFAM'
plt.figure(figsize=(20, 6))
sns.boxplot(x='RFAM', y='recovery_avg', data=taro_df_new)
plt.title("Recovery Rate by RFAM family ID")
plt.xlabel("RFAM family ID")
plt.ylabel("Recovery Rate")
plt.show()
```
